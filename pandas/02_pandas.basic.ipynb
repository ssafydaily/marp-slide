## 딕셔너리에서 DataFrame 생성

```python
df = pd.DataFrame(
    {
        "Name": [
            "홍길동",
            "김싸피",
            "고길동",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)
print(df)
print(df.index)
print(df.columns)
```

```python
df.describe()
```

- Number인 열 선택

```python
print(df.select_dtypes('number'))
print(type(df.select_dtypes('number')))
```

- 객체 자료형 선택

```python
print(df.select_dtypes('object'))
print(type(df.select_dtypes('object')))
```
