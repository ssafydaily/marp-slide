```python
# 객체의 비교 연산
src = [1, 2, 3]
dst = src

print(src, dst)
print(src == dst)
print(id(src), id(dst))
print(src is dst)
```

```python
# 객체의 비교 연산
src = [1, 2, 3]
dst = src
src[0] = -10
print(src, dst)
print(src == dst)
print(id(src), id(dst))
print(src is dst)
```

```python
# 객체의 비교 연산
src = [1, 2, 3]
dst = src[:]
# src[0] = -10
print(src, dst)
print(src == dst)
print(id(src), id(dst))
print(src is dst)
```

```python
from copy import deepcopy
# 객체의 비교 연산
src = [[1, 2], [3, 4]]
dst = src[:]
src[0][0] = -10
print(src, dst)
print(src == dst)
print(src is dst)
```

```python
from copy import deepcopy
# 객체의 비교 연산
src = [[1, 2], [3, 4]]
dst = deepcopy(src)

print(src, dst)
print(src == dst)
print(src is dst)
```
